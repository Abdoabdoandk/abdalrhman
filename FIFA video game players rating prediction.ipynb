using BeautifulSoup library to extract all the data for the first 100 player from fifa website

```python
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
import requests
import html5lib 
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split , GridSearchCV
from sklearn import metrics
from sklearn import preprocessing
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import Ridge
url = requests.get('https://www.ea.com/games/ea-sports-fc/ratings').text
soup = BeautifulSoup(url, 'html5lib')
html_tables = soup.find_all('table')

names_list = []
Position_list = []
OVR_list = []
PAC_list = []
SHO_list = []
PAS_list = []
DRI_list = []
DEF_list = []
PHY_list = []

def extract_name(x):
    table_1 = html_tables[0].find_all('tr')[x]
    name_cell = table_1.find_all('td')[1].find('a')
    names_list.append(name_cell.text)
    
def extract_position(x):
    table_1 = html_tables[0].find_all('tr')[x]
    position_cell = table_1.find_all('td')[4].find('a')
    Position_list.append(position_cell.text)

def extract_OVR(x):
    table_1 = html_tables[0].find_all('tr')[x]
    OVR_cell = list(table_1.find_all('td')[5].stripped_strings)
    OVR_list.append(OVR_cell[1])

def extract_PAC(x):
    table_1 = html_tables[0].find_all('tr')[x]
    PAC_cell = list(table_1.find_all('td')[6].stripped_strings)
    PAC_list.append(PAC_cell[1])

def extract_SHO(x):
    table_1 = html_tables[0].find_all('tr')[x]
    SHO_cell = list(table_1.find_all('td')[7].stripped_strings)
    SHO_list.append(SHO_cell[1])  

def extract_PAS(x):
    table_1 = html_tables[0].find_all('tr')[x]
    PAS_cell = list(table_1.find_all('td')[8].stripped_strings)
    PAS_list.append(PAS_cell[1])

def extract_DRI(x):
    table_1 = html_tables[0].find_all('tr')[x]
    DRI_cell = list(table_1.find_all('td')[9].stripped_strings)
    DRI_list.append(DRI_cell[1])

def extract_DEF(x):
    table_1 = html_tables[0].find_all('tr')[x]
    DEF_cell = list(table_1.find_all('td')[10].stripped_strings)
    DEF_list.append(DEF_cell[1])

def extract_PHY(x):
    table_1 = html_tables[0].find_all('tr')[x]
    PHY_cell = list(table_1.find_all('td')[11].stripped_strings)
    PHY_list.append(PHY_cell[1])

for x in range(1,400,4):
    extract_name(x)
    extract_position(x)
    extract_OVR(x)
    extract_PAC(x)
    extract_SHO(x)
    extract_PAS(x)
    extract_DRI(x)
    extract_DEF(x)
    extract_PHY(x)


print(names_list)
print(Position_list)
print(OVR_list)
print(PAC_list)
print(SHO_list)
print(PAS_list)
print(DRI_list)
print(DEF_list)
print(PHY_list)


```

Pace table

```python
Acceleration = []
for row in html_tables[0].find_all('tr')[4:]:
    table_1 = row.find('td').find_all('div')
    for x in range(len(table_1)):
        y = list(table_1[x].stripped_strings)
        if y and y[0] == 'Acceleration' :
         Acceleration.append(y[1])
Acceleration = Acceleration[0:100] 

SprintSpeed = []
for row in html_tables[0].find_all('tr')[4:]:
    table_1 = row.find('td').find_all('div')
    for x in range(len(table_1)):
        y = list(table_1[x].stripped_strings)
        if y and y[0] == 'Sprint Speed' :
         SprintSpeed.append(y[1])
SprintSpeed = SprintSpeed[0:100]  

Dict_PAC = {'Name':names_list  ,'Position':Position_list, 'Acceleration':Acceleration , 'Sprint Speed': SprintSpeed, "PAC":PAC_list}
dfPAC = pd.DataFrame(Dict_PAC)
dfPAC = dfPAC.set_index(pd.RangeIndex(start=1, stop=len(dfPAC) + 1))
dfPAC.head()
```

Turn position column to numrical for modeling

```python
dfPAC[['Acceleration','Sprint Speed','PAC']] = dfPAC[['Acceleration','Sprint Speed','PAC']].astype(int)
x = pd.get_dummies(dfPAC['Position']).astype(int)
dfPAC1 = pd.concat([dfPAC,x],axis=1)
dfPAC1.drop(columns='Position')
dfPAC1.head()
```

create the model that predict the overall Pace depend on the attributes in the pace section

```python
x_PAC = dfPAC1.drop(columns=['PAC','Name','Position']).values
y_PAC = dfPAC1['PAC'].values
x_PAC = preprocessing.StandardScaler().fit(x_PAC).transform(x_PAC)
x_train_PAC , x_test_PAC, y_train_PAC , y_test_PAC = train_test_split(x_PAC , y_PAC , test_size=0.1, random_state=3)
model_PAC = Ridge(alpha=0.1).fit(x_train_PAC,y_train_PAC)
yhat = model_PAC.predict(x_test_PAC)
mse = mean_squared_error(y_test_PAC,yhat)
accuracy = metrics.r2_score(y_test_PAC,yhat)
print('The mean square error of this model is',mse)
print('The accuracy of this model is',accuracy*100,'%')
yhat_PAC  =model_PAC.predict(x_PAC)
for x in range(0,len(yhat_PAC)-1):
    yhat_PAC[x] = math.ceil(yhat_PAC[x])
yhat_PAC = yhat_PAC.astype(int)
```
